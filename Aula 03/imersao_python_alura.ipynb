```python
import pandas as pd
import plotly.express as px
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
df_principal = pd.read_excel("tabeladeacoes.xlsx", sheet_name="Principal")
df_principal
```

```python
df_totalacoes = pd.read_excel("tabeladeacoes.xlsx", sheet_name="Total_de_acoes")
df_totalacoes
```

```python
df_ticker = pd.read_excel("tabeladeacoes.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("tabeladeacoes.xlsx", sheet_name="ChatGPT")
df_chatgpt
```

```python
df_principal = df_principal[["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)" : "valor_final", "Var. Dia (%)" : "var_dia_pct"}).copy()
df_principal
```

```python
df_principal["var_pct"] = df_principal["var_dia_pct"]/100
df_principal["valor_inicial"] = df_principal["valor_final"]/(1 + df_principal["var_pct"])
df_principal
```

```python
df_principal = df_principal.merge(df_totalacoes, left_on="Ativo", right_on="Código", how="left")
df_principal
```

```python
df_principal = df_principal.drop(columns={"Código"})
df_principal
```

```python
df_principal["variacao_rs"] = (df_principal["valor_final"]-df_principal["valor_inicial"]) * df_principal["Qtde. Teórica"]
df_principal
```

```python
df_principal["Qtde. Teórica"] = df_principal["Qtde. Teórica"].astype(int)
df_principal = df_principal.rename(columns={"Qtde. Teórica" : "qtd_teorica"}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on="Ativo", right_on="Ticker", how="left")
df_principal = df_principal.drop(columns=["Ticker"])
df_principal
```

```python
df_chatgpt = df_chatgpt.rename(columns={"Nome":"nome_empresa"}).copy()
df_chatgpt
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on="Nome", right_on="nome_empresa", how="left")
df_principal = df_principal.drop(columns=["nome_empresa"])
df_principal
```

```python
df_principal['cat_idade'] = df_principal['Idade'].apply(lambda x: "Mais de 100 anos" if x > 100 else ("Menos de 50 anos" if x < 50 else "Entre 50 e 100 anos"))
df_principal
```

```python
# Calculando o maior valor, o menor valor e a média da coluna 'variacao_rs'
maior_valor = df_principal['variacao_rs'].max()
menor_valor = df_principal['variacao_rs'].min()
media_valor = df_principal['variacao_rs'].mean()

# Calculando a média da coluna 'variacao_rs' para cada categoria 'subiu', 'estável' e 'desceu'
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()
media_estavel = df_principal[df_principal['resultado'] == 'Estável']['variacao_rs'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()

# Imprimindo os resultados
print("Maior\tR$", maior_valor)
print("Menor\tR$", menor_valor)
print("Média\tR$", media_valor)
print("Média de quem subiu\tR$", media_subiu)
print("Média de quem desceu\tR$", media_desceu)
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmentos = df_principal_subiu.groupby("Segmento")["variacao_rs"].sum().reset_index()
df_analise_segmentos
```

```python
df_analise_saldo = df_principal.groupby("resultado")["variacao_rs"].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x = "resultado", y = "variacao_rs", text="variacao_rs", title = "variação por resultado")
#desafio: alterar formatação dos números.
fig.update_layout(yaxis_tickprefix="R$")
fig.update_traces(texttemplate="R$ %{text:,.2f}", textposition="inside")
fig.show()
```

```python
#desafio: gerar grafico pizza do df_analise_segmentos.
graf_pizza_segmentos = px.pie(df_analise_segmentos, values="variacao_rs", names="Segmento", title="Variação em reais por Segmento")
graf_pizza_segmentos.show()
```

```python
#desafio fazer groupby da faixa etária
df_analise_faixaetaria = df_principal.groupby("cat_idade").size().reset_index(name='qtd_empresas')
df_analise_faixaetaria
```

```python
#desafio grafico de pizza do df_analise_faixaetaria
graf_pizza_idades = px.pie(df_analise_faixaetaria, values="qtd_empresas", names="cat_idade", title="Quantidade de empresas por faixa etária")
graf_pizza_idades.show()
```
